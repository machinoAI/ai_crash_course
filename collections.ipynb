```python
from collections import Counter, defaultdict, deque, namedtuple, OrderedDict, ChainMap, UserList
from pprint import pprint
```

# 1. Counter

```python
nums = [10,20,30,40,50, 20]
count = Counter(nums)
print(count)
```

```python
list(count.elements())
```

```python
count.most_common(2)
```

# 2. defaultdict

```python
d = defaultdict(list)

d["A"].append(1)
d
```

```python
defaultdict(int)    # 0
defaultdict(float)  # 0.0
defaultdict(set)    # set()
```

```python
nums = [10,20,30,40,50, 60]
```

```python
groups  = defaultdict(list)
for x in nums:
    groups[x%4].append(x)
        
pprint(dict(groups), width=30)
```

# 3. deque

```python
q = deque([1,2,3])
q.append(4)
q
```

```python
q.appendleft(0)
q
```

```python
q.pop()
q
```

```python
q.popleft()
q
```

```python
q.rotate()
q
```

```python
q.extend([10,20])
q
```

# 4. namedtuple

```python
Student = namedtuple(
    "Student",
    ["name", "age", "marks"]
    
)

s = Student(
    "Bony",
    25,
    90
)

print(s.name)
print(s.age)
print(s.marks)
```

```python
#Convert to dictionary
print(s._asdict())
```

```python
#Create from iterable
data = ["Alex", 25, 90]

s = Student(*data)
s
```

```python
#Replace values
s2 = s._replace(age=30)
s2
```

# 5. OrderedDict

```python
d = OrderedDict()

d["a"] = 1
d["b"] = 2

print(d)
```

```python
d.move_to_end("a")
d
```

```python
d.popitem(last=False)
d
```

# 6. ChainMap

```python
d1 = {"a": 1}
d2 = {"b": 2}

c = ChainMap(d1, d2)
c
```

```python
print(c["a"])
print(c["b"])
```

# 7. UserList

```python
class MyList(UserList):

    def append(self, item):
        print("Adding:", item)
        super().append(item)

l = MyList()

l.append(10)
```
