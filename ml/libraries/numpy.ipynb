```python
arr1 = np.array([[1,2,3],
                 [6,7,8]])
```

```python
arr1
```

```python
arr1.shape
```

```python
arr1.reshape(3,2)
```

```python
flat = arr1.flatten()
flat 
```

```python
flat[0] = 50
print(flat)
print(arr1)  # Original array didn't get modify
```

```python
# creates a view instead of copy
r = arr1.ravel()  
print(r)
```

```python
r[0] = 100
print(r)
print(arr1)  # original array got modified
```

```python
print(arr1)
print("transposed array:", arr1.transpose())
```

```python
# T: same as transpose()
print(arr1)
print("transposed array:",arr1.T)
```
