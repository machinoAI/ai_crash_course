```python
#Check whether a year is a leap year.
def leap_year(year):
    if year % 400 ==0:
        print(f"{year} is a leap year")
        
    elif year % 100 ==0:
        print(f"{year} is not a leap year")
         
    elif year % 4 ==0:
        print(f"{year} is a leap year")
    else:
        print(f"{year} is not a leap year")
```

```python
leap_year(2000)
```

```python
#Group records by month.
records = [
    {"id": 1, "date": "17-07-2026", "amount": 100},
    {"id": 2, "date": "20-07-2026", "amount": 200},
    {"id": 3, "date": "05-08-2026", "amount": 150},
    {"id": 4, "date": "10-08-2026", "amount": 300},
    {"id": 5, "date": "12-09-2026", "amount": 250}
]
from collections import defaultdict
grouped = defaultdict(list)

for record in records:
    dt = datetime.strptime(record["date"], "%d-%m-%Y")
    month_key = dt.strftime("%d-%Y")
    grouped[month_key].append(record)

print(grouped)
```

```python
#Sort objects by timestamp.
event1 = [
    {
        "id": 1,
        "event": "login",
        "timestamp": "2026-07-17 09:15:30"
    },
    {
        "id": 2,
        "event": "purchase",
        "timestamp": "2026-07-15 14:45:10"
    },
    {
        "id": 3,
        "event": "logout",
        "timestamp": "2026-07-17 18:20:05"
    },
    {
        "id": 4,
        "event": "signup",
        "timestamp": "2026-07-10 08:00:00"
    },
    {
        "id": 5,
        "event": "password_reset",
        "timestamp": "2026-07-16 11:30:45"
    }
]

event2 = [
    {
        "id": 1,
        "event": "login",
        "timestamp": 1784260530
    },
    {
        "id": 2,
        "event": "purchase",
        "timestamp": 1784097910
    },
    {
        "id": 3,
        "event": "logout",
        "timestamp": 1784292605
    },
    {
        "id": 4,
        "event": "signup",
        "timestamp": 1783651200
    }
]
```

```python
#Find the first Monday of a month.

month = 7
year = 2026

def first_monday(month, year):
    current_date = datetime(year, month, 1)

    while current_date.weekday() !=0:
        current_date +=timedelta(days=1)
        
    return current_date

first_monday(month, year)
```
